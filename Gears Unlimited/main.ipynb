<h1>Inventory Replenishment Policy </h1>

<p>Gears unlimited distributes specialty gears, derailleurs, and brakes for
high-end mountain and BMX bikes. One of their most steady selling item is the PK35 derailleur. They sell about 1500 of the PK35â€™s a year. They
cost $75 each to procure from a supplier and Gears Unlimited assumes that the cost of capital is 20% a year. It costs about $350 to place and receive an order of the PK35s, regardless of the quantity of the order.
</p>


 <h4> - How many PK35s should Gear Unlimited order at a time to minimize
the average annual cost in terms of purchase cost, ordering costs and
holding cost?
</h4>

- Demand = $1500$ $ units/year
- Cost of procurement = $75$ $/unit
- Cost of capital = $0,2$ $ per year
- Ordering cost = $350$ $/order

Q = Order quantity
<p> </p>
Q* = Optimal order quantity

What is the object function?

- Total cost = Purchase cost + Ordering cost + Holding cost

What is the purchase cost?

- Purchase cost = Demand * Cost of procurement
- Purchase cost =  $1500$ $ units/year * $75$ $/unit

What is the ordering cost?

- Ordering cost = (Demand / Cost of Ordering) * Quantity ordered
- Ordering cost = $(350*1500)/Q$

What is the holding cost?

- Holding cost = (Q / 2) * Cost of capital
- Holding cost = $(Q/2)*0.2$

TotalCost(Q) = $1500*75 + (350*1500)/Q + (Q/2)*0.2$
=> $112500 + 525000/Q + 0.1Q$

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
def purchase_cost(x):
    return 1500 * 75

def orderCost(x):
    return (350)*(150)/x

def holding_cost(x):
    return (x/2)*75*0.2

def OrderHolding(x):
    return orderCost(x) + holding_cost(x)

print("Here are all the functions for the costs")

x = np.arange(8, 500, 2)

# Plot
plt.plot(x, orderCost(x), label='Order Cost')
plt.plot(x, holding_cost(x), label='Holding Cost')
plt.plot(x, OrderHolding(x), label='Order Cost + Holding Cost')

plt.xlabel('Order Quantity (Q)')
plt.ylabel('Cost $')
plt.grid(True)
plt.legend()

plt.show()
```

1. Determine the Objective Function:
- TotalCost(Q) = $112500 + 525000/Q + 7.5Q$
2. Determine the Derivative of the Objective Function:
- TotalCost'(Q) = $-525000/Q^2 + 7.5$
3. Set the Derivative of the Objective Function to Zero:
- $-525000/Q^2 + 7.5 = 0$
- $-525000/Q^2 = -7.5$
- $Q^2 = 70000$
- $Q = \sqrt{70000}$
- $Q = 264.57$
4. Determine the Second Derivative of the Objective Function:
- TotalCost''(Q) = $1050000/Q^3$
- Because the second derivative is positive, the function is concave up at Q = 264.57.
- we know that this Q* is the minimum of the function.
- because the second derivative is positive, the function is concave up at Q = 264.57.

<h4>Therefore, the optimal order quantity is 264.57 units.</h4>



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Data
def purchase_cost():
    return 1500 * 75  # This is a constant, doesn't depend on x, so no need for x as an argument

def order_cost(x):
    return (350)*(1500)/x  # Order cost as a function of x

def holding_cost(x):
    return (x/2)*75*0.2  # Holding cost as a function of x

# Numeric TotalCost function
def total_cost_numeric(x):
    return purchase_cost() + order_cost(x) + holding_cost(x)

# Symbol x for symbolic calculations
x_sym = sp.symbols('x')

# Symbolic TotalCost function
total_cost_sym = purchase_cost() + order_cost(x_sym) + holding_cost(x_sym)

# Derive the TotalCost function symbolically
total_cost_derivative = sp.diff(total_cost_sym, x_sym)

# Find the minimum extreme point for the TotalCost function symbolically
extreme_points = sp.solve(total_cost_derivative, x_sym)

# Evaluate extreme points numerically
extreme_points_numeric = [float(point.evalf()) for point in extreme_points]

# Generate x values for plotting
x_values = np.linspace(1, 500, 500)  # Avoid starting at 0 to prevent division by zero
print(f"Here is the TotalCost function: {total_cost_sym}")
print(f"Here is the derivative of the TotalCost function: {total_cost_derivative}")
print(f"Extreme points: {extreme_points_numeric}")

# Plot the Total Cost function for numeric x values
plt.plot(x_values, [total_cost_numeric(x) for x in x_values], label='Total Cost')

# Plot the minimum cost point
for point in extreme_points_numeric:
    if 1 <= point <= 500:
        plt.plot(point, total_cost_numeric(point), 'ro', label='Minimum Cost')
    else:
        print(f"Extreme point {point} is outside the range of x, and is irrelevant for this case bacause it is not a minimum")

plt.xlabel('Order Quantity (Q)')
plt.ylabel('Cost $')
plt.title('Total Cost vs. Order Quantity')
plt.grid(True)
plt.legend()
plt.show()
```
