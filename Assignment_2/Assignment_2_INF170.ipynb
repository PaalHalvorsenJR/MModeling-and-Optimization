<h2>Task 1</h2>

### Sets 
- $P$: Plants
- $S$: Suppliers

### Parameters  

- $c_p$: Cost of plant $p \in P$
- $C_p$: Capacity of plant $p \in P$
- $yi_p$: Yield per unit of plant $p \in P$
- $d_{sp}$: Distance from supplier $s \in S$ to plant $p \in P$
- $u_{sp}$: Cost per unit from supplier $s \in S$ to plant $p \in P$
- $t_{sp}$: Truck cost from supplier $s \in S$ to plant $p \in P$
- $tc_{sp}$: Truck capacity
- $s_s$: Supply of supplier $s \in S$


### Decision variables

- $x_{sp} \geq 0$: Biomass transported from supplier $s$ to plant $p$ $\quad \forall s \in S, \forall p \in P$
- $y_p \in {0,1}$: Binary variable, 1 if plant $p$ is open, 0 otherwise $\quad \forall p \in P$
- $z_{sp} \geq 0, z_{sp} \in \mathbb{N}$: Number of trucks from supplier $s$ to plant $p$ $\quad \forall s \in S, \forall p \in P$

### Objective function

$\text{min} \sum\limits_{p \in P} c_p \cdot y_p + \sum\limits_{s \in S, p \in P} (u_{sp} \cdot x_{sp} + t_{sp} \cdot z_{sp})$

### Constraints

- $\text{Suppy\_Constraint: } : \sum\limits_{p \in P} x_{sp} \leq s_s \quad \forall s \in S$
<br>
- $\text{Demand\_Constraint: } \sum\limits_{s \in S, p \in P} x_{sp} \cdot yi_p \cdot y_p \geq 500000000 \quad $
<br>
- $\text{Truck\_Capacity: } x_{sp} \leq z_{sp} \cdot tc_{sp} \quad \forall s \in S, \forall p \in P$
<br>
<br>
- $\text{Plant\_Capacity: } \sum\limits_{s \in S} x_{sp} \leq C_p \cdot y_p \quad  \forall p \in P $

<h2>Task 2</h2>

### Sets 
- $P$: Plants
- $H$: Hubs
- $S$: Suppliers

### Parameters
#### Suppliers
- $s_s$: Supply of supplier $s \in S$

#### Hubs
- $ch_h$: Cost of hub $h \in H$
- $Ch_h$: Capacity of hub $h \in H$
- $dh_{sh}$: Distance from supplier $s \in S$ to hub $h \in H$
- $uh_{sh}$: Cost per unit from supplier $s \in S$ to hub $h \in H$
- $th_{sh}$: Truck cost from supplier $s \in S$ to hub $h \in H$
- $tch_{sh}$: Truck capacity from supplier $s \in S$ to hub $h \in H$

#### Plants
- $cp_p$: Cost of plant $p \in P$
- $Cp_p$: Capacity of plant $p \in P$
- $yi_p$: Yield per unit of plant $p \in P$
- $dp_{hp}$: Distance from hub $h \in H$ to plant $p \in P$
- $up_{hp}$: Cost per unit from hub $h \in H$ to plant $p \in P$
- $tp_{hp}$: Train cost from hub $h \in H$ to plant $p \in P$
- $tcp_{hp}$: Train capacity from hub $h \in H$ to plant $p \in P$ 

### Decision variables
- $x_{sh} \geq 0$: Biomass transported from supplier $s$ to hub $p$ $\quad \forall s \in S, \forall h \in H$
- $y_{hp} \geq 0$: Biomass transported from hub $h$ to plant $p$ $\quad \forall h \in H, \forall p \in P$

- $z_{h} \in {0,1}$: Binary variable, 1 if hub $h$ is open, 0 otherwise $\quad \forall h \in H$
- $w_{p} \in {0,1}$: Binary variable, 1 if plant $p$ is open, 0 otherwise $\quad \forall p \in P$

- $trucks_{sh} \geq 0, trucks_{sh} \in \mathbb{N}$: Number of trucks from supplier $s$ to hub $h$ $\quad \forall s \in S, \forall h \in H$
- $trains_{hp} \geq 0, trains_{hp} \in \mathbb{N}$: Number of trains from hub $h$ to plant $p$ $\quad \forall h \in H, \forall p \in P$

### Objective function
$\text{min} \sum\limits_{p \in P} w_p \cdot cp_p + \sum\limits_{h \in H} z_h * ch_h + \sum\limits_{s \in S, h \in H} (x_{sh} \cdot uh_{sh} \cdot dh_{sh} + th_{sh} \cdot trucks_{sh}) + \sum\limits_{h \in H, p \in P} (y_{hp} \cdot up_{hp} \cdot dp_{hp} + tp_{hp} \cdot trains_{hp})$

### Constraints
- $\text{Suppy\_Constraint: } : \sum\limits_{h \in H} x_{sh} \leq s_s \quad \forall s \in S$

- $\text{Demand\_Constraint: } \sum\limits_{s \in S, h \in H} y_{hp} \cdot yi_p \cdot w_p \geq 500,000,000 \quad $

- $\text{Truck\_Capacity: }  x_{sh} \leq trucks_{sh} \cdot tch_{sh} \quad \forall s \in S, \forall h \in H$

- $\text{Train\_Capacity: }  y_{hp} \leq trains_{hp} \cdot tcp_{hp} \quad \forall h \in H, \forall p \in P$

- $\text{Hub\_Capacity: } \sum\limits_{s \in S} x_{sh} \leq Ch_h \cdot z_h \quad  \forall h \in H$

- $\text{Plant\_Capacity: } \sum\limits_{h \in H} y_{hp} \leq Cp_p \cdot w_p \quad  \forall p \in P$

- $\text{Flow\_Balance: } \sum\limits_{s \in S} x_{sh} = \sum\limits_{ p \in P} y_{hp} \quad  \forall h \in H $

#### thirdparty 
- $third = 2000$: Cost of third party transport

### Decision variables
- $x_{sh} \geq 0$: Biomass transported from supplier $s$ to hub $p$ $\quad \forall s \in S, \forall h \in H$
- $y_{hp} \geq 0$: Biomass transported from hub $h$ to plant $p$ $\quad \forall h \in H, \forall p \in P$

- $\text{Plant\_Capacity: } \sum\limits_{h \in H} y_{hp} \leq Cp_p \cdot w_p \quad \forall p \in P$

- $\text{Flow\_Balance: } \sum\limits_{s \in S} x_{sh} + bio\_th_{h} = \sum\limits_{ p \in P} y_{hp} \quad \forall h \in H$
