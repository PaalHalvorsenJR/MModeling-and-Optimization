-- We are manufacturing a product where we know

- The cost function = $f(made) = 500 000 + 75x$
- The Demand function = $f(price) = 20 000 - 80p$
- Profit function = $f(Demand, price) = p(20 000 - 80p)- [500 000 + 75(20 000 -80p)]$

- And therefore the profit function = $f(price) = -80p^2 + 26 000p - 2 000 000$

-- We want to find the price that will maximize the profit.


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the profit function
def profit_function(p):
    return -80*p**2 + 26000*p - 2000000

# Generate x values
x = np.linspace(125, 200, 200)

# Generate y values

# Calculate y values using the profit function
y = profit_function(x)

# Plot the profit function
plt.plot(x, y)
plt.xlabel('Price of Item')
plt.ylabel('Total Profit')
plt.title('Profit vs Price')
plt.grid(True)
plt.show()
```

- We find the price p that maximizes the profit function by first finding the derivative of the profit function and then setting it to zero.

- The derivative of the profit function is $f'(price) = -160p + 26 000$
- Setting the derivative to zero, we get $-160p + 26 000 = 0$
- Solving for p, we get $p = 162.5$

Now we can find the maximum profit by substituting the value of p into the profit function.
$ f(162.5) = -80(162.5)^2 + 26 000(162.5) - 2 000 000$
- we get => $ f(162.5) = 112 500$
